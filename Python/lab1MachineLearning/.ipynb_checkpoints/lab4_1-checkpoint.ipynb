```python
import pandas as pd
import numpy as np
from sklearn.ensemble import BaggingClassifier, GradientBoostingRegressor, AdaBoostClassifier, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pylab as plt
from sklearn.naive_bayes import GaussianNB
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('D:/FAX/mashin_ljorning/Lab_4_Ensemble/glass.csv', index_col=0)
print(data.head())
X_train, X_test, y_train, y_test = train_test_split(data.values[::, :-1], data.values[::, -1], test_size=0.33, random_state=0)
```

```python
scores1 = np.array([])
scores2 = np.array([])
n_estimators = np.array(range(1, 152, 5))

for i in n_estimators:
    classifier = BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth=5), n_estimators=i, random_state=0).fit(X_train, y_train)
    scores1 = np.append(scores1, accuracy_score(y_train, classifier.predict(X_train)))
    scores2 = np.append(scores2, accuracy_score(y_test, classifier.predict(X_test)))

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(n_estimators, scores1)
ax1.set_xlabel('models count')
ax1.set_ylabel('accuracy')
ax1.set_title('train')
ax2.plot(n_estimators, scores2)
ax2.set_title('test')
ax2.set_xlabel('models count')
ax2.set_ylabel('accuracy')
```

```python
scores1 = np.array([])
scores2 = np.array([])
n_estimators = np.array(range(10, 151, 10))

for i in n_estimators:
    classifier = BaggingClassifier(base_estimator=KNeighborsClassifier(), n_estimators=i, random_state=0).fit(X_train, y_train)
    scores1 = np.append(scores1, accuracy_score(y_train, classifier.predict(X_train)))
    scores2 = np.append(scores2, accuracy_score(y_test, classifier.predict(X_test)))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 8))
ax1.plot(n_estimators, scores1)
ax1.set_xlabel('models count')
ax1.set_ylabel('accuracy')
ax1.set_title('train')
ax2.plot(n_estimators, scores2)
ax2.set_title('test')
ax2.set_xlabel('models count')
ax2.set_ylabel('accuracy')
```

```python
data = pd.read_csv('D:/FAX/mashin_ljorning/Lab_4_Ensemble/titanic.csv', index_col=0)
data.head()
```

```python
sns.heatmap(data.isnull(), cbar = False).set_title("Missing values heatmap")
```

```python
data.Embarked.fillna(data.Embarked.mode()[0], inplace = True)
data.Cabin.fillna('NA', inplace = True)
data.Age.fillna(data.Age.mean(), inplace = True)
```

```python
from sklearn.preprocessing import OrdinalEncoder

embarkedEncoder = OrdinalEncoder()
data.Embarked = embarkedEncoder.fit_transform(np.array(data.Embarked).reshape(-1, 1))

sexEncoder = OrdinalEncoder()
data.Sex = sexEncoder.fit_transform(np.array(data.Sex).reshape(-1, 1))

nameEncoder = OrdinalEncoder()
data.Name = nameEncoder.fit_transform(np.array(data.Name).reshape(-1, 1))

cabinEncoder = OrdinalEncoder()
data.Cabin = cabinEncoder.fit_transform(np.array(data.Cabin).reshape(-1, 1))

data = data.drop(columns=['Name', 'Ticket'])
data.head()
```

```python
X = data.values[::, 1:]
y = data.values[::, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
sns.heatmap(data.corr(), cmap= sns.color_palette(palette="RdBu"),linewidths=0.5,annot=True)
```

```python
from sklearn.ensemble import StackingClassifier

estimators = [('randomforest', RandomForestClassifier(n_estimators=5, random_state=0))]
clf = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())

clf.fit(X_train, y_train).score(X_test, y_test)
```
