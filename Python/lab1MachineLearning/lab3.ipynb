```python
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('D:\FAX\mashin_ljorning/Lab_3_Clustering/pluton.csv')
train_data = data.values
scores = np.array([])
max_iters = np.array(range(1, 20))

for i in max_iters:
    model = KMeans(n_clusters=3, random_state=6, max_iter=i).fit(train_data)
    print('not scaled iters count', model.n_iter_)
    scores = np.append(scores, silhouette_score(train_data, model.labels_))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(max_iters, scores, 'ro')
ax1.set_title('not scaled')
ax1.set_xlabel('max_iters')
ax1.set_ylabel('scores')
print('scores: ', scores)

train_data = data.values
scores = np.array([])
max_iters = np.array(range(1, 20))
train_data_scaled = StandardScaler().fit_transform(train_data)

for i in max_iters:
    model = KMeans(n_clusters=3, random_state=6, max_iter=i).fit(train_data_scaled)
    print('scaled iters count', model.n_iter_)
    scores = np.append(scores, silhouette_score(train_data_scaled, model.labels_))
    
ax2.plot(max_iters, scores, 'ro')
ax2.set_title('scaled')
ax2.set_xlabel('max_iters')
ax2.set_ylabel('scores')
print('scores: ',scores)
```

```python
from sklearn.datasets import load_sample_image
from sklearn.utils import shuffle
from skimage.io import imread

image = imread('img.jpg')
image = np.array(image, dtype=np.float64) / 255
fig, ax = plt.subplots(figsize=(25, 15))
ax.imshow(image)

w, h, d = original_shape = tuple(image.shape)
image_array = np.reshape(image, (w * h, d))

image_array_sample = shuffle(image_array, random_state=0)[:1000]
kmeans = KMeans(n_clusters=16, random_state=0).fit(image_array_sample)

labels = kmeans.predict(image_array)
```

```python
new_image = np.zeros((w, h, d))
label_idx = 0
for i in range(w):
    for j in range(h):
        new_image[i][j] = kmeans.cluster_centers_[labels[label_idx]]
        label_idx += 1
fig, ax = plt.subplots(figsize=(25, 15))
ax.imshow(new_image)
```
